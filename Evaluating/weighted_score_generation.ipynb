```python
import pandas as pd
```

```python
scores_df = pd.read_csv('code_bleu_evaluation.csv')
```

```python
scores_df.head()
```

```python
from codebleu import calc_codebleu
```

```python
def equal_weighted_scores(reference, prediction):
    return calc_codebleu([reference], [prediction], lang="python", weights=(0.25, 0.25, 0.25, 0.25), tokenizer=None)
```

```python
def equal_weighted_codebleu_score(reference, prediction):
    return equal_weighted_scores(reference, prediction)['codebleu']
```

```python
def add_codebleu_scores_for_before_and_after(scores_df, column_name):
    for indx in range(len(scores_df)):
        scores_df[column_name][indx] = equal_weighted_codebleu_score(scores_df['before_merge'][indx], scores_df['after_merge'][indx])
```

```python
def add_codeblue_scores_column(scores_df, column_name):
    scores_df[column_name] = None
    add_codebleu_scores_for_before_and_after(scores_df, column_name)
```

```python
BEFORE_AFTER_COLUMN = 'before_after_score'
```

```python
# Drop column in already Present
scores_df = scores_df.drop([BEFORE_AFTER_COLUMN], axis = 1)
```

```python
add_codeblue_scores_column(scores_df, BEFORE_AFTER_COLUMN)
```

```python
scores_df.head()
```

```python
def add_codebleu_weight_column(scores_df):
    scores_df['codebleu_weight'] = abs(scores_df['before_after_score'] - 1)
```

```python
scores_df.head()
```

```python
add_codebleu_weight_column(scores_df)
```

```python
scores_df.head()
```

```python
def add_weighted_codebleu_score_for_column(scores_df, column_name):
    scores_df['weighted_codebleu_' + column_name] = scores_df[column_name] * scores_df['codebleu_weight']
```

```python
add_weighted_codebleu_score_for_column(scores_df, 'codeT5_secure_before_score')
```

```python
scores_df = scores_df.drop(['weighted_codebleucodeT5_secure_before_score'], axis = 1)
```

```python
add_weighted_codebleu_score_for_column(scores_df, 'codeT5_secure_after_score')
```
